```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("howpop_train.csv")
```

```python
df.shape
```

```python
df.head(3).T
```

```python
df.drop(filter(lambda c: c.endswith("_lognorm"), df.columns),axis=1, inplace=True)
```

```python
df.describe().T
```

```python
df.describe(include=["bool","object"]).T
```

```python
df['published']=pd.to_datetime(df.published, yearfirst=True) #привод строки в 
```

```python
#Создадим несколько столбцов на основе данных о времени публикации:

df["year"] = [d.year for d in df.published]
df["month"] = [d.month for d in df.published]
df["dayofweek"] = [d.isoweekday() for d in df.published]
df["hour"] = [d.hour for d in df.published]
df.head().T
```

```python
pd.crosstab(df['month'],df['year'])
```

```python
#1. В каком месяце (и какого года) было больше всего публикаций?
ax = sns.countplot( x = 'year',hue = 'month',data = df,)
ax.legend(loc = 'upper left', fontsize= '8')
```

```python
#2. Проанализируйте публикации в месяце из предыдущего вопроса
ax = sns.countplot( x = df[(df['month'] == 3) & (df['year'] == 2015)]['dayofweek'], hue = df['domain'])
ax.legend(loc = 'upper left', fontsize= '8')
```

```python
#3. Когда лучше всего публиковать статью?
#число просмотров в среднем по времени
ax = sns.boxplot(x = 'hour', y = 'views', data = df)
ax.set_ylim(0,20000)
```

```python
#число комментариев в среднем по времени
ax = sns.boxplot(x = 'hour', y = 'comments', data = df)
ax.set_ylim(0,100)
```

```python
#число комментариев на гиктаймсе в зависимости от времени
ax= sns.boxplot(x = df[df['domain'] == 'geektimes.ru']['hour'], y = df['comments'])
ax.set_ylim(0,100)
```

```python
#4. Кого из топ-20 авторов чаще всего минусуют?
top_authors = df['author'].value_counts().head(20).index
print(top_authors)
ax = sns.boxplot(x = 'author', y = 'votes_minus', data=df[df['author'].isin(top_authors)],)
ax.set_ylim(0,30)
plt.xticks(rotation=90)
```

```python
#5. Сравните субботы и понедельники. 
#Правда ли, что по субботам авторы пишут в основном днём, а по понедельникам — в основном вечером?
df[df['dayofweek'] == 6]['hour'].median(),\
df[df['dayofweek'] == 1]['hour'].median()
```
